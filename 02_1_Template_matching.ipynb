```
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```
!wget -O original https://raw.githubusercontent.com/MLman/computer-vision-AAA534/master/02-edge/mario.jpg
!wget -O template https://raw.githubusercontent.com/MLman/computer-vision-AAA534/master/02-edge/mario_coin.jpg
```

```
img_rgb = cv2.imread('original')
img_rgb.shape
```

```
plt.imshow(img_rgb)
```

```
original = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2RGB)
original.shape
```

```
plt.imshow(original)
```

```
template = cv2.imread('template')
template = cv2.cvtColor(template, cv2.COLOR_BGR2RGB)
r, c, _ = template.shape
template.shape
```

```
plt.figure(figsize = (1,1))
plt.imshow(template)
plt.show()
```

```
res = cv2.matchTemplate(original, template, cv2.TM_CCOEFF_NORMED)
```

```
plt.imshow(res)
plt.show()
```

```
threshold = 0.8
loc = np.where(res >= threshold) # (row, column)
for pt in zip(*loc[::-1]): # (column, row)
    cv2.rectangle(original, (pt[0], pt[1]), (pt[0] + c, pt[1] + r), (0,255,0), 2)
```

```
plt.imshow(original)
plt.show()
```
